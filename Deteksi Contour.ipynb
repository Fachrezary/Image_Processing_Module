```python
import cv2
import matplotlib.pyplot as plt

def crop_image_with_contours(image_path):
    # Load image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to obtain binary image
    _, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Find contours in the binary image
    contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a figure and axes
    fig, ax = plt.subplots()

    # Display the original image
    ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # Crop and display the image with contours
    for contour in contours:
        # Approximate the contour to simplify its shape
        epsilon = 0.01 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Find the bounding box of the contour
        x, y, w, h = cv2.boundingRect(approx)

        # Crop the image using the bounding box coordinates
        cropped_image = image[y:y+h, x:x+w]

        # Display the cropped image
        ax.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))

    # Set the aspect ratio to 'equal'
    ax.set_aspect('equal')

    # Show the plot
    plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt

def plot_image_with_contours(image_path):
    # Load image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to obtain binary image
    _, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Find contours in the binary image
    contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a figure and axes
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # Display the original image
    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Original Image')

    # Draw contours on the image
    for contour in contours:
        # Approximate the contour to simplify its shape
        epsilon = 0.01 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Draw the contour itself
        axes[1].plot(approx[:, 0, 0], approx[:, 0, 1], 'r', linewidth=2)

    # Set the aspect ratio to 'equal'
    axes[1].set_aspect('equal')
    axes[1].set_title('Image with Contours')

    # Show the plot
    plt.tight_layout()
    plt.show()
```
